```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
def plt_imshow(image):
    title=''
    # convert the image frame BGR to RGB color space and display it
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.axis(False)
    plt.imshow(image)
    plt.title(title)
    plt.grid(True)
    plt.axis('on')
    plt.show()
```

```
img = cv2.imread('/work/cat2.jpg')
plt_imshow(img)
```

```
M = np.ones(img.shape, dtype=np.uint8)*25
img_t = cv2.add(img, M)
```

```
plt_imshow(img_t)
```

```
(r, g, b) = cv2.split(img)
```

```
plt_imshow(r)
plt_imshow(g)
plt_imshow(b)
```

```
zeros = np.zeros(img.shape[:2], dtype="uint8")
plt_imshow(cv2.merge([r, zeros, zeros]))
plt_imshow(cv2.merge([zeros, g, zeros]))
plt_imshow(cv2.merge([zeros, zeros, b]))
plt_imshow(cv2.merge([b, g, r]))
```

```
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```
plt_imshow(thresh1)
```

```
img2 = cv2.imread('/work/cat.jpg')
ret, thresh2 = cv2.threshold(img2, 177, 255, cv2.THRESH_BINARY)
```

```
plt_imshow(thresh2)
```
