```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
img_c = cv.imread('./cat.jpg')
```

```
img = cv.cvtColor(img_c, cv.COLOR_BGR2GRAY)
```

```
def showimg(img):
    plt.imshow(img, "gray")
showimg(img)
```

```
kernel = np.ones((5,5), np.uint8)
```

# Morphological Operations
    1. Erosion
    2. Dilation

```
img_erosion = cv.erode(img, kernel, iterations=1)
img_dilation = cv.dilate(img, kernel, iterations=1)
```

```
showimg(img_erosion)
```

```
showimg(img_dilation)
```

# Morphological Operations on Noisy image
    1. Opening
    2. Closing
    3. Morphological Gradient
    4. Black Hat
    5. Top Hat

```
img[200:300, 200:300] = 0
```

```
showimg(img)
```

```
kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
showimg(opening)
```

```
kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
closing = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
showimg(closing)
```

```
kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
showimg(gradient)
```

```
kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)
showimg(blackhat)
```

```
kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
showimg(tophat)
```
