```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Edge Detection/gates.png')
plt.imshow(img)
```

```python
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow(img, cmap='Greys')
```

## Reducing image noise using blurring

```python
# Reduce image noise

# median blur
img_m = cv2.medianBlur(img, 3)
# gaussian blur
img_g = cv2.GaussianBlur(img, (5,5), 0)
# average blur
img_a = cv2.blur(img, (5,5))
```

```python
plt.imshow(img_m)
```

```python
plt.imshow(img_g)
```

```python
plt.imshow(img_a)
```

## Edge detection

```python
blurred = img_m
big = cv2.Canny(blurred, 10, 200)
mid = cv2.Canny(blurred, 30, 150)
low = cv2.Canny(blurred, 240, 250)

fig = plt.figure(figsize=(20, 20))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(big, cmap='Greys')

fig.add_subplot(rows, columns, 2)
plt.imshow(mid, cmap='Greys')

fig.add_subplot(rows, columns, 3)
plt.imshow(low, cmap='Greys')
```

```python
blurred = img_g
big = cv2.Canny(blurred, 10, 200)
mid = cv2.Canny(blurred, 30, 150)
low = cv2.Canny(blurred, 240, 250)

fig = plt.figure(figsize=(20, 20))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(big, cmap='Greys')

fig.add_subplot(rows, columns, 2)
plt.imshow(mid, cmap='Greys')

fig.add_subplot(rows, columns, 3)
plt.imshow(low, cmap='Greys')
```
